### 4. Если имеете супруга или партнёра, как долго вы живёте вместе (в годах)?

```python
set(df[defs.QUESTION_4].tolist())
```

```python
def remove_no_digits(string):
    digits = ''
    for char in string:
        if char.isdigit() or char == '.':
            digits += char
    return float(digits)

def convert_string(cell_content):
    if isinstance(cell_content, str):
        result = remove_no_digits(cell_content)
        if 'месяц' in cell_content:
            result /= 12
    else:
        result = cell_content
    return result
```

```python
df.replace({
    defs.QUESTION_4:
    {
        '-':np.NaN,
        '?':np.NaN
    }
}, inplace=True)

df[defs.QUESTION_4] = df[defs.QUESTION_4].apply(convert_string)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(defs.QUESTION_4, fontsize=16)

for label, ax in zip(defs.LABELS, axes):
    a = df[df['Метка'] == label][defs.QUESTION_4].isnull().sum()
    b = df[df['Метка'] == label][defs.QUESTION_4].notnull().sum()
    wedges, _, _ = ax.pie([a, b], autopct='%1.1f%%')
    ax.set_title(label)
    if ax is axes[-1]:
        ax.legend(wedges, set(('нет супруга/партнёра', 'есть супруг/партнёр')), bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
fig.suptitle(defs.QUESTION_4, fontsize=16)
axes[0].hist(df[df['Метка'] == defs.LABELS[0]][defs.QUESTION_4].tolist(), bins=60, range=(0, 60))
axes[0].set_title(defs.LABELS[0])
axes[0].set_xlim(0, 65)

axes[1].hist(df[df['Метка'] == defs.LABELS[1]][defs.QUESTION_4].tolist(), bins=60, range=(0, 60))
axes[1].set_title(defs.LABELS[1])
axes[1].set_xlim(0, 65)

axes[2].hist(df[df['Метка'] == defs.LABELS[2]][defs.QUESTION_4].tolist(), bins=60, range=(0, 60))
axes[2].set_title(defs.LABELS[2])
axes[2].set_xlim(0, 65)

plt.show()
```

### Кодирование и приведение к типам, доступным для конвертации DataFrame в Dataset

```python
df.replace({defs.QUESTION_4:{np.NaN:0}}, inplace=True)
df.replace({defs.QUESTION_7:{np.NaN:'-'}}, inplace=True)
df.replace({defs.QUESTION_22:{np.NaN:0}}, inplace=True)
df.replace({defs.QUESTION_24:{np.NaN:0}}, inplace=True)
```

```python
def encode_categorical_columns(dataframe):
    for column_name in defs.CATEGORICAL_COLUMN_NAMES.keys():
        # Столбец в DataFrame становится "категориальным". Это означает, что ячейки в нём могут
        # принимать только значения из допустимых категорий либо np.NaN. Допустимые категории
        # определяются при создании такого столбца либо множеством различных ячеек, либо
        # специальным списком. Если допустимые категории определяются специальным списком и 
        # значение ячейки не входит в этот список, её значение становится np.NaN.
        dataframe[column_name] = pd.Categorical(dataframe[column_name], categories=defs.CATEGORICAL_COLUMN_NAMES[column_name])
        # Категории кодируются числами 0, 1, 2, ... np.NaN числом -1.
        dataframe[column_name] = dataframe[column_name].cat.codes
        # Приведение к типу, допустимому для конвертации DataFrame в Dataset.
        dataframe[column_name] = dataframe[column_name].astype('int64')
```
