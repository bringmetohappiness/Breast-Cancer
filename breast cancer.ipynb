```python
def autolabel(rects, labels=None, height_factor=1.01):
    for i, rect in enumerate(rects):
        height = rect.get_height()
        if labels is not None:
            try:
                label = labels[i]
            except (TypeError, KeyError):
                label = ' '
        else:
            label = '%d' % int(height)
        ax.text(rect.get_x() + rect.get_width()/2., height_factor*height,
                '{}'.format(label),
                ha='center', va='bottom')


fig, ax = plt.subplots()
ax.bar(LABELS, [NORM_N_ROWS, BENIGN_TUMOR_N_ROWS, MALIGNANT_TUMOR_N_ROWS], tick_label=LABELS)
plt.xticks(rotation='vertical')
fig.suptitle('Распределение точек данных по классам')
autolabel(ax.patches, height_factor=0.85)
plt.show()
```

```python
def plot_pies(question):
    fig, axes = plt.subplots(1, 3, figsize=(18, 10))
    
    for label, ax in zip(LABELS, axes):
        sizes = []
        answer_set = set(dataframe[question].tolist())
        for answer in answer_set:
            amount = dataframe[(dataframe['Метка'] == label) & (dataframe[question] == answer)].shape[0]
            sizes.append(amount)
        ax.pie(sizes, labels=answer_set)
        ax.set_title(label)

    plt.show()


plot_pies(QUESTION_3)
```

### Вопрос 4: Если имеете супруга или партнера, как долго вы живете вместе (в годах)?

```python
set(dataframe[QUESTION_4].tolist())
```

Необходимо избавиться от 'лет' и 'годов' и перевести в float.  
Для значений '-' и '?' пока придумал присвоить им 0.  

```python
dataframe.replace(
    {
        QUESTION_4:
        {
            '-':0,
            '?':0
        }
    }, inplace=True
)


def rm_nodigits(cell_content):
    if isinstance(cell_content, str):
        digits = str()
        for char in cell_content:
            if char.isdigit() or char == '.':
                digits += char
        result = float(digits)
    else:
        result = cell_content
    
    return result


dataframe[QUESTION_4] = dataframe[QUESTION_4].apply(rm_nodigits)
```
